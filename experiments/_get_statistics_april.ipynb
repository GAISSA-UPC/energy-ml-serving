```python
# List to hold DataFrames
dfs = []

# Iterate over all CSV files in specified directories
for file_path in glob.glob('results_march/*/*.csv', recursive=True):
    df = pd.read_csv(file_path)
    dfs.append(df)

# Concatenate all DataFrames
results_df = pd.concat(dfs, ignore_index=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
models
```

```python
model_name = models[0]

#torch_df = results_df[results_df['engine'] == 'none']
torch_df = results_df[(results_df['engine'] == 'none') & (results_df['model'] == model_name)]
onnx_df = results_df[(results_df['engine'] == 'onnx') & (results_df['model'] == model_name)]
torchscript_df = results_df[(results_df['engine'] == 'torchscript') & (results_df['model'] == model_name)]
ov_df = results_df[(results_df['engine'] == 'ov') & (results_df['model'] == model_name)]
```

```python
properties = ['duration','cpu_energy','ram_energy','energy_consumed',]
```

```python
#property = 'duration'
for property in properties:
    # Create a 2x2 grid of subplots
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
    
    
    # Create boxplots for each variable in the corresponding subplot
    sns.boxplot(y=property, data=torch_df, ax=axes[0, 0])
    sns.boxplot(y=property, data=onnx_df, ax=axes[0, 1])
    sns.boxplot(y=property, data=ov_df, ax=axes[1, 0])
    sns.boxplot(y=property, data=torchscript_df, ax=axes[1, 1])
    
    # Adding titles to each subplot
    axes[0, 0].set_title('Torch')
    axes[0, 1].set_title('ONNX')
    axes[1, 0].set_title('OpenVINO')
    axes[1, 1].set_title('TorchScript')
    
    
    # Set plot title and adjust layout
    fig.suptitle(f'Boxplots of {property} using {model_name}') #energy consumed [kWh]
    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust the position of the title
    
    # Show the plot
    plt.show()
```

```python
property = 'energy_consumed'

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Create boxplots for each variable in the corresponding subplot
sns.boxplot(y=property, data=torch_df, ax=axes[0, 0])
sns.boxplot(y=property, data=onnx_df, ax=axes[0, 1])
sns.boxplot(y=property, data=ov_df, ax=axes[1, 0])
sns.boxplot(y=property, data=torchscript_df, ax=axes[1, 1])

# Adding titles to each subplot
axes[0, 0].set_title('Torch')
axes[0, 1].set_title('ONNX')
axes[1, 0].set_title('OpenVINO')
axes[1, 1].set_title('TorchScript')

min_y = 0.6e-8
max_y = 1e-8
# Set y-axis limits for each subplot
axes[0, 0].set_ylim(min_y, max_y)  # Adjust the range as needed
axes[0, 1].set_ylim(min_y, max_y)  # Adjust the range as needed
axes[1, 0].set_ylim(min_y, max_y)  # Adjust the range as needed
axes[1, 1].set_ylim(min_y, max_y)  # Adjust the range as needed

# Set plot title and adjust layout
fig.suptitle(f'Boxplots of {property} using {model_name}') #energy consumed [kWh]
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust the position of the title

# Show the plot
plt.show()
```
